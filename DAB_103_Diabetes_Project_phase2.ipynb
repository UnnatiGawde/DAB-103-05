# Cleaning Diabetes dataset

# To remove the null values in diabetes dataset

```python
clean_df = df.dropna(subset=['Age'])
clean_df
```

```python
null_counts = clean_df.isnull().sum()
print(null_counts)
```

```python
# check for duplicate rows
duplicate_rows = clean_df[clean_df.duplicated()]
print("Number of duplicate rows:", len(duplicate_rows))
```

```python
# check for duplicates in each column
for col in clean_df.columns:
    if duplicate_rows[col].any():
        print("Duplicate values found in column:", col)
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Q1) What is the relationship between stress, family history of diabetes, and the development of diabetes in individuals?

```python
# filter by diabetic people
stress_df = clean_df[(clean_df['Diabetic'] == 'yes')]

# Create pivot table
pivot_table_stress_df = pd.pivot_table(stress_df,
                                       values= ['Diabetic'],
                                       index=["Family_Diabetes", "Stress"],
                                       aggfunc=len)

# Convert pivot table to table
pivot_table_stress_df = pivot_table_stress_df.reset_index()
pivot_table_stress_df.columns.name = None
pivot_table_stress_df

# create barplots for each sub-dataframe in each subplot
sns.barplot(data = pivot_table_stress_df, x = 'Family_Diabetes', y = 'Diabetic', hue = 'Stress')
plt.xlabel('Family_Diabetes')
plt.ylabel('Diabetic')
plt.title('Diabetic patients with family history of diabetes with Stress')
plt.show()

# display the plot
plt.show()
```

```python
# Convert pivot table to table
pivot_table_stress_df = pivot_table_stress_df.reset_index()
pivot_table_stress_df.columns.name = None
pivot_table_stress_df
```

#Q2) What are the standards used to classify people as having diabetes, and how are they classified according to their gender and the duration of their exercise?

```python
# filter by diabetic people
activity_df = clean_df[(clean_df['Diabetic'] == 'yes')]

# Create pivot table
pivot_table_activity_df = pd.pivot_table(activity_df,
                                       values= ['Diabetic'],
                                       index=["Gender", "PhysicallyActive"],
                                       aggfunc=len)

# Convert pivot table to table
pivot_table_activity_df = pivot_table_activity_df.reset_index()
pivot_table_activity_df.columns.name = None
pivot_table_activity_df

# create barplots for each sub-dataframe in each subplot
sns.barplot(data = pivot_table_activity_df, x = 'Gender', y = 'Diabetic', hue = 'PhysicallyActive')
plt.xlabel('Gender')
plt.ylabel('Diabetic')
plt.title('Number of Diabetic People by Gender and Physical Activity Status')

# Move the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# display the plot
plt.show()
```

# Q4) What is the distribution of diabetic patients by Age group and has family history of Diabetes?

```python
diabetic_family_history = clean_df[(clean_df['Diabetic'] == 'yes') & (clean_df['Family_Diabetes'] == 'yes')]
diabetic_family_history_by_age = diabetic_family_history.groupby('Age')['Age'].count().sort_values(ascending=True)

# plot the bar chart
ax = diabetic_family_history_by_age.plot(kind='bar', cmap='GnBu_r')

# add labels and title
ax.set_xlabel('Age group')
ax.set_ylabel('Number of patients')
ax.set_title('Diabetic patients with family history of diabetes across age groups')

# add number labels on top of each bar
for i, v in enumerate(diabetic_family_history_by_age):
    ax.text(i, v+1, str(v), ha='center', fontsize=10)

# display the plot
plt.show()
```

# Q5) How is regular medicine intake related to diabetic patients

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# filter by diabetic people
diabetic_df = clean_df[clean_df['Diabetic'] == 'yes']

# create a contingency table using pandas crosstab function
cont_table = pd.crosstab(diabetic_df['Diabetic'], diabetic_df['RegularMedicine'])

# create a heatmap using seaborn
sns.heatmap(cont_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Relationship between Diabetic and RegularMedicine')
plt.xlabel('Diabetic')
plt.ylabel('Count')
plt.show()
```
